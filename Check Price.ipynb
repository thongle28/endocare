```python
import sqlite3
import pandas as pd
import sources.parts_list_0802 as pl
```

```python
conn = sqlite3.connect('history.db')
```

```python
a = pl.parts_list()
b = pl.summary_report()
```

```python
plf = a.part_list_final('FMSV2022040081',conn)
```

```python
rma_list = '''

FMSV2023090222
FMSV2023090195
FMSV2023080018
FMSV2023080016
FMSV2023110022
FMSV2023080075
FMSV2023080015
FMSV2023080074
FMSV2023110031

'''
rma_list = list(rma_list.strip().split('\n'))
```

```python
rma = 'FMSV2023090222'
plf = a.part_list_final(rma,conn)
ffvn_price = plf['FFVN Price'].sum()
plf['Dealer Price'].sum()
```

```python
rma_price = {}
for rma in rma_list:
    try:
        plf = a.part_list_final(rma,conn)
        ffvn_price = plf['FFVN Price'].sum()
        dealer_price = plf['Dealer Price'].sum()
        issue = b.issues(conn,rma)
        rma_price.update({rma:(ffvn_price,dealer_price,issue)})
    except Exception as e:
        print(rma, e)
    
    
```

```python
b.issues(conn,'FMSV2022040081')
```

```python
import xlwings as xw
wb = xw.Book()
sh  = wb.sheets('Sheet1')
sh.range('A1').value = 'RMA'
sh.range('B1').value = 'FFVN'
sh.range('C1').value = 'Dealer'
sh.range('D1').value = 'Issue'
```

```python
i = 0
for rma,price in rma_price.items():
    sh.range('A2').offset(i,0).value = rma
    sh.range('B2').offset(i,0).value = price[0]
    sh.range('C2').offset(i,0).value = price[1]
    sh.range('D2').offset(i,0).value = price[2]
    i +=1
print('Done')
```

```python
wb.save('inspection/RMa_price_2301130.xlsx')
```

```python
wb.close()
```
