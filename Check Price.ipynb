```python
import sqlite3
import pandas as pd
import sources.parts_list_0802 as pl
```

```python
conn = sqlite3.connect('history.db')
```

```python
a = pl.parts_list()
b = pl.summary_report()
```

```python
plf = a.part_list_final('FMSV2022040081',conn)
```

```python
rma_list = '''
FMSV2019080079
FMSV2019070031
FMSV2019100025
FMSV2019100018
FMSV2021100018
FMSV2021060017
FMSV2022040081
FMSV2022050043
FMSV2022080072
FMSV2022090010
FMSV2022090028
FMSV2022100091
FMSV2022090065
FMSV2022110058
FMSV2022120038
FMSV2022120043
FMSV2023020002
FMSV2023090043
FMSV2023040070
FMSV2023060018
FMSV2023060027
FMSV2023060068
FMSV2023050045L
FMSV2023070046
FMSV2023070056
FMSV2023080095
FMSV2023090026




'''
rma_list = list(rma_list.strip().split('\n'))
```

```python
rma = 'FMSV2022120042'
a.part_list_final(rma,conn)
```

```python
rma_price = {}
for rma in rma_list:
    try:
        plf = a.part_list_final(rma,conn)
        ffvn_price = plf['FFVN Price'].sum()
        dealer_price = plf['Dealer Price'].sum()
        issue = b.issues(conn,rma)
        rma_price.update({rma:(ffvn_price,dealer_price,issue)})
    except Exception as e:
        print(rma, e)
    
    
```

```python
b.issues(conn,'FMSV2022040081')
```

```python
import xlwings as xw
wb = xw.Book()
sh  = wb.sheets('Sheet1')
sh.range('A1').value = 'RMA'
sh.range('B1').value = 'FFVN'
sh.range('C1').value = 'Dealer'
sh.range('D1').value = 'Issue'
```

```python
i = 0
for rma,price in rma_price.items():
    sh.range('A2').offset(i,0).value = rma
    sh.range('B2').offset(i,0).value = price[0]
    sh.range('C2').offset(i,0).value = price[1]
    sh.range('D2').offset(i,0).value = price[2]
    i +=1
```

```python
wb.save('inspection/RMa_price_230925.xlsx')
```
