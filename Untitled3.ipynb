```python
import pandas as pd
from sqlite3 import connect

conn = connect('bachmai.db')
```

```python
q = '''
        SELECT DISTINCT model,count(*) AS QTY
        
        
        
        FROM install
        WHERE [customer code] = 'FMSV00038'
        GROUP BY model
        
        ORDER BY QTY DESC
        
    '''
pd.read_sql(q,conn)
```

```python
q = '''
        SELECT DISTINCT model,
        strftime('%Y',[date installed]) as YEAR, COUNT(*)
        
        
        
        FROM install
        WHERE [customer code] = 'FMSV00038'
        GROUP BY model, YEAR
        
       
        
    '''
details = pd.read_sql(q,conn)
```

```python
details.to_excel('details.xlsx',index = False)
```
