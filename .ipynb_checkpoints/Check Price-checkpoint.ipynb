```python
import sqlite3
import pandas as pd
import sources.parts_list_0705 as pl
```

```python
conn = sqlite3.connect('history.db')
```

```python
a = pl.parts_list()
```

```python
plf = a.part_list_final('FMSV2023040057',conn)
```

```python
dealer_price = plf['Dealer_AMOUNT'].sum()
```

```python
ffvn_price = plf['FFVN_AMOUNT'].sum()
```

```python
rma_list = '''
FMSV2023060041
FMSV2023060042
FMSV2023060043
FMSV2023060044
FMSV2023060045
FMSV2023060046
FMSV2023060047
FMSV2023060048
FMSV2023020034
FMSV2023050025
FMSV2023060058
FMSV2023030060

'''
rma_list = list(rma_list.strip().split('\n'))
```

```python
rma = 'FMSV2022120042'
a.part_list_final(rma,conn)
```

```python
rma_price = {}
for rma in rma_list:
    try:
        plf = a.part_list_final(rma,conn)
        ffvn_price = plf['FFVN Price'].sum()
        dealer_price = plf['Dealer Price'].sum()
        rma_price.update({rma:(ffvn_price,dealer_price)})
    except Exception as e:
        print(rma, e)
    
    
```

```python
rma_price
```

```python
import xlwings as xw
wb = xw.Book()
```

```python
sh  = wb.sheets('Sheet1')
sh.range('A1').value = 'RMA'
sh.range('B1').value = 'FFVN'
sh.range('C1').value = 'Dealer'
```

```python
i = 0
for rma,price in rma_price.items():
    sh.range('A2').offset(i,0).value = rma
    sh.range('B2').offset(i,0).value = price[0]
    sh.range('C2').offset(i,0).value = price[1]
    i +=1
```

```python
wb.save('RMa_price_230721.xlsx')
```

```python
wb.close()
```
