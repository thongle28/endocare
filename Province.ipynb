```python
key_list={}
```

```python
found = 0
not_found = 0

for i in range(len(cus)):
    key = ''
    code = cus['No.'][i]
    customer_name = cus['Name'][i]
    
    addr1 = cus['Addr1'][i]
    addr2 = cus['Addr2'][i]
    city = cus['City'][i]
    bingo = False
    try:
        addr1 = addr1.upper()
    except:
        pass
    try:
        addr2 = addr2.upper()
    except:
        pass
    try:
        city = city.upper()
    except:
        pass
    for cit in cities:
#         if cit in customer_name.upper() or cit in addr1 or cit in addr2 or cit in city:
        if cit in customer_name.upper():
            print(f'{code}, {customer_name}, key: {cit}')
            bingo = True
            found +=1
            key_list.update({code:[customer_name,cit]})
            break
        try:
            if cit in addr1.upper():
                print(f'{code}, {customer_name}, key: {cit}')
                bingo = True
                found +=1
                key_list.update({code:[customer_name,cit]})
                break
        except:
            pass
        
        try:
            if cit in addr2.upper():
                print(f'{code}, {customer_name}, key: {cit}')
                bingo = True
                found +=1
                key_list.update({code:[customer_name,cit]})
                break
        except:
            pass
        
        try:
            if cit in city.upper():
                print(f'{code}, {customer_name}, key: {cit}')
                bingo = True
                found +=1
                key_list.update({code:[customer_name,cit]})
                break
        except:
            pass
        
    if not bingo:
        print(f'{customer_name} city not found.')
        not_found+=1
```
