```python
#combine receive
# rma_list = tuple(rma_record)
q = f'''
        SELECT  DISTINCT c.[rma no.],c.customer_name,c.model,c.serial_no,
        strftime('%Y-%m-%d',CASE 
            WHEN receive NOT NULL THEN receive
            ELSE c.recieve_date END) AS [Receive Date (IN)],
        
        CASE
            WHEN m.quotation IS NULL THEN c.in_inspect_date
            WHEN m.quotation IN ('FOC','WARRANTY') THEN c.in_inspect_date
            ELSE m.quotation  END as [Quoted Date],
        
        confirmation AS [PO rec. Date],
        m.part_order_date as [Part Order Date],
        CASE
            WHEN m.part_received_date NOT NULL THEN m.part_received_date
            ELSE c.[Part Select Date] END AS [Part Receive Date],
        ROUND(JULIANDAY(m.note) - JULIANDAY(c.[Part Select Date])) AS [TAT Parts],
        
        CASE 
            WHEN m.start_repair NOT NULL THEN m.start_repair
            ELSE strftime('%Y-%m-%d',re.[Start Time]) END AS[Repair Start Date],
        ROUND(JULIANDAY(c.qc) - JULIANDAY(re.[Start Time])) AS [TAT Repair],
        CASE
            WHEN m.qc_date NOT NULL THEN m.qc_date
            ELSE strftime('%Y-%m-%d',c.qc) END AS [Ship Date (OUT)],
        
        ROUND(JULIANDAY(c.qc) - JULIANDAY(receive)) AS [TAT Days Completed]
        
        FROM (consolidated c
        LEFT JOIN repair_code re ON c.[RMA NO.] = re.[RMA NO.])
        LEFT JOIN new_ml m ON c.[RMA NO.] = m.rma
        WHERE c.[rma no.] IN {rma_list}
        
        ORDER BY c.[rma no.]
        
    '''

tat_repair = pd.read_sql(q,conn)
tat_repair.head()#.to_excel('inspection\\tat_repair.xlsx',index=False)
```

```python
for i in range(len(tat_repair)):
    if tat_repair['Quoted Date'][i]== None:
        tat_repair['Quoted Date'][i] = tat_repair['Receive Date (IN)'][i]
    if tat_repair['PO rec. Date'][i]== None:
        tat_repair['PO rec. Date'][i] = tat_repair['Quoted Date'][i]
    if tat_repair['Part Order Date'][i]== None:
        tat_repair['Part Order Date'][i] = tat_repair['PO rec. Date'][i]
    if tat_repair['Part Receive Date'][i]== None:
        tat_repair['Part Receive Date'][i] = tat_repair['Part Order Date'][i]
    if tat_repair['Repair Start Date'][i]== None:
        tat_repair['Repair Start Date'][i] = tat_repair['Part Receive Date'][i]
    if tat_repair['Ship Date (OUT)'][i]== None:
        tat_repair['Ship Date (OUT)'][i] = tat_repair['Repair Start Date'][i]
tat_repair
```

```python
# amount of waiting confirm
q = '''SELECT COUNT(*),SUM(price) AS [Amount of Waiting confirm($)]
        FROM m_list
        WHERE status LIKE '%waiting confirm'
    '''
pd.read_sql(q,conn)
```

```python
# total price
q = '''
        SELECT COUNT(*),SUM(part_list_date) AS [Amount of Transfers($)]
        FROM transfers
        WHERE return is null
        AND customer NOT LIKE '%FUJIFILM%'
    '''
pd.read_sql(q,conn)
```

```python
# min receive
q = '''SELECT MIN(RECEIVE)
        from transfers
        WHERE return is null
        AND part_list_date not null
        '''
pd.read_sql(q,conn)
```

```python
# GROUP BY CUSTOMER
q = '''
        SELECT customer,round(sum(part_list_date),0) AS [Price($)]
        FROM TRANSFERS
        WHERE return is null
        AND part_list_date NOT NULL
        AND customer NOT LIKE '%FUJIFILM%'
        
        group by customer
        HAVING [price($)] >10000
        ORDER BY [price($)] DESC
    '''
pd.read_sql(q,conn)
```

```python
#count BY customer
q = '''SELECT customer,round(sum(part_list_date),0) AS [Price($)], count(*) AS [count]
        
        FROM transfers
        
        WHERE return is null
        AND part_list_date NOT NULL
        AND customer NOT LIKE '%FUJIFILM%'
        
        GROUP BY customer
        HAVING [count] >=3
        ORDER BY [Price($)] DESC,[count] DESC
        

'''

pd.read_sql(q,conn)
```

```python
model ='EG-600WR'

q = f'''SELECT rma,tf.model,sn,strftime("%d-%m-%Y",receive) as received,
                strftime("%Y",[date installed]) as[Year Installed],

                issue,round(part_list_date) AS [Dealer Price (GP 41%)],
                round(part_list_date*0.85) AS [GP 20%]
        
        FROM transfers tf
        LEFT JOIN install ins ON tf.sn = ins.[serial no.]
        
        WHERE return is null
        AND part_list_date NOT NULL
        AND customer = '{customer_name}'
        and tf.model = '{model}'
        ORDER BY [Year Installed] DESC, [Dealer Price (GP 41%)] DESC
    '''
print(f'{customer_name.upper()} - Model: {model}')
pd.read_sql(q,conn)
```
