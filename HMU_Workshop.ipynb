```python
import pandas as pd
from sqlite3 import connect
```

```python
conn = connect('hmu.db')
```

```python
q = '''
        SELECT i.model,i.[serial no.],i.[group],i.type,i.[last repair rma],i.[Current Status Name],
        c.[approval],c.[repair size],c.approval
        
        FROM install i
        LEFT JOIN consolidated c ON i.[last repair rma] = c.[rma no.]
        WHERE i.[Customer Code] = 'FMSV00079'
'''
pd.read_sql(q,conn)
```
