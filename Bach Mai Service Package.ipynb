```python
import pandas as pd
from sqlite3 import connect

conn = connect('history.db')
```

```python
# create RMA List for Bach Mai
q = '''
        SELECT *
        FROM consolidated c
        WHERE customer_code = 'FMSV00038'
        AND repair_status IN ('Inspection','Parts Selection')
   '''
rma_list = pd.read_sql(q,conn) 
(rma_list['RMA No.'])
```

```python
# Select from parts
q = f'''
        SELECT p.[rma no.] ,p.part_description,
        pn.short_name
        
        FROM parts p
        LEFT JOIN part_name pn ON p.part_description = pn.part_description
        WHERE [RMA NO.] in {tuple(rma_list['RMA No.'])}
        
    '''
short_name_parts = pd.read_sql(q,conn)
short_name_parts
```
